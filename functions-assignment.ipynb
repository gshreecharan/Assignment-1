```python
# 6
def uppercase_if_string(name):
   if type(name)== str:
     print(name.upper())
   else:
     print("invalid input")
uppercase_if_string(input())
```

```python
# 7
def safe_divide(num,den):
 if den!=0:
     print(num/den)
 else:
     print("cannot divide")
safe_divide(20,2)
```
